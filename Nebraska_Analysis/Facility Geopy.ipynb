```python
import pandas as pd
from geopy.geocoders import Nominatim
import ast
```

```python
# Set up the geocoder
geolocator = Nominatim(user_agent="my_geocoder")

# Define a function to get county name from lat_long
def get_county_name(lat_long_str):
    if lat_long_str:
        lat_long = ast.literal_eval(lat_long_str)
        location = geolocator.reverse((lat_long[0], lat_long[1]), exactly_one=True)
        if location and 'address' in location.raw:
            county = location.raw['address'].get('county')
            return county
    return None
```
